```python
import csv

# 使用csv模組逐行讀取，只讀取檔案頭和少量資料來獲取基本資訊
def get_csv_info(file_path, max_rows=5):
    rows_count = 0
    with open(file_path, 'r', encoding='utf-8') as f:
        csv_reader = csv.reader(f)
        header = next(csv_reader)  # 讀取表頭
        columns_count = len(header)
        
        # 讀取少量行來取得樣本
        sample_rows = []
        for i, row in enumerate(csv_reader):
            if i < max_rows:
                sample_rows.append(row)
            rows_count += 1
            
            # 每讀取100萬行更新進度
            if rows_count % 1000000 == 0:
                print(f"已讀取 {rows_count} 行...")
                
    return {
        'columns_count': columns_count,
        'rows_count': rows_count,
        'header': header,
        'sample_rows': sample_rows
    }

# 執行分析
info = get_csv_info('training.csv')
print(f"總行數: {info['rows_count']}")
print(f"欄位數: {info['columns_count']}")
```

```python
import csv

# 設定檔案路徑
csv_file_path = 'E:\\Tbrain_stock_analysis\\training.csv'
output_txt_path = 'column_names.txt'

# 只讀取欄位名稱（第一列）
with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)

# 將欄位名稱寫入 txt 檔
with open(output_txt_path, 'w', encoding='utf-8') as txtfile:
    for column_name in header:
        txtfile.write(column_name + '\n')

print(f"已成功將欄位名稱寫入 {output_txt_path}")
```
